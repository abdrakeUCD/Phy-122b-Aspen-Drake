```python
# -----------------------------
# HELPER FUNCTIONS: spectrum loading
# -----------------------------

def load_spectrum(filepath: Path):
    """
    Returns:
      channels (np.ndarray): integer channel bin centers [0..N-1]
      counts   (np.ndarray): integer counts per channel
    Supports:
      (A) event-list: one column of ADC channels (one per event)
      (B) histogram: two columns (channel, counts)
      (C) histogram: one row/col counts array
    """
    filepath = Path(filepath)
    arr = np.genfromtxt(filepath, delimiter=None)

    if arr.ndim == 1:
        # Either event-list (values are channels) OR 1D histogram counts.
        # Heuristic: if max is "large-ish" compared to length, assume event list.
        if arr.size > 0 and np.nanmax(arr) > arr.size:
            # Event list
            events = arr[~np.isnan(arr)].astype(int)
            nchan = int(events.max()) + 1
            counts = np.bincount(events, minlength=nchan)
            channels = np.arange(nchan)
            return channels, counts
        else:
            # 1D histogram counts
            counts = np.nan_to_num(arr).astype(float)
            channels = np.arange(len(counts))
            return channels, counts

    if arr.ndim == 2 and arr.shape[1] >= 2:
        ch = arr[:, 0]
        ct = arr[:, 1]
        # If channels are not integer (or start not at 0), still OK
        return ch.astype(float), ct.astype(float)

    raise ValueError(f"Unrecognized spectrum format: {filepath}")
```

```python
# -----------------------------
# HELPER FUNCTIONS: background
# -----------------------------

def subtract_background(sample_ch, sample_ct, bg_ch, bg_ct,
                        t_sample_s=15*60, t_bg_s=15*60,
                        t_sample_err_s=30, t_bg_err_s=30):
    """
    Subtract background assuming both are measured for known durations.
    Works in rates, then converts back to expected counts over sample duration.

    Returns:
      corrected_counts, corrected_err (1σ)
    """
    # Ensure matching channel grids
    # If they differ, interpolate bg onto sample channels (counts vs channel).
    if not np.array_equal(sample_ch, bg_ch):
        bg_ct_interp = np.interp(sample_ch, bg_ch, bg_ct, left=0.0, right=0.0)
    else:
        bg_ct_interp = bg_ct

    # Rates
    r_s = sample_ct / t_sample_s
    r_b = bg_ct_interp / t_bg_s

    # Poisson count uncertainties
    # σ(count) ~ sqrt(count). Rate σ: sqrt(N)/t
    sig_rs_stat = np.sqrt(np.maximum(sample_ct, 0.0)) / t_sample_s
    sig_rb_stat = np.sqrt(np.maximum(bg_ct_interp, 0.0)) / t_bg_s

    # Time uncertainty propagation (small, but include)
    sig_rs_time = np.abs(sample_ct) * (t_sample_err_s / (t_sample_s**2))
    sig_rb_time = np.abs(bg_ct_interp) * (t_bg_err_s / (t_bg_s**2))

    sig_rs = np.sqrt(sig_rs_stat**2 + sig_rs_time**2)
    sig_rb = np.sqrt(sig_rb_stat**2 + sig_rb_time**2)

    r_corr = r_s - r_b
    sig_rcorr = np.sqrt(sig_rs**2 + sig_rb**2)

    # Convert back to expected counts over sample duration
    ct_corr = r_corr * t_sample_s
    sig_ct_corr = sig_rcorr * t_sample_s

    return ct_corr, sig_ct_corr


def plot_spectrum(ch, ct, title, savepath=None, ylog=False):
    plt.figure()
    plt.plot(ch, ct, linewidth=1)
    plt.title(title)
    plt.xlabel("ADC channel")
    plt.ylabel("Counts")
    if ylog:
        plt.yscale("log")
    if savepath is not None:
        plt.savefig(savepath, dpi=200, bbox_inches="tight")
    plt.show()
    
```

```python
# -----------------------------
# HELPER FUNCTIONS: gaussian stuff
# -----------------------------

def gaussian(x, A, mu, sigma, C):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2)) + C


def find_peak_gaussian(ch, ct, approx_center, approx_width,
                       use_poisson_weights=True,
                       save=False, filename="peak_fit.png"):
    """
    Fit a Gaussian in a window around approx_center.
    Returns:
      mu_hat, mu_err, sigma_hat, A_hat
    """
    ch = np.asarray(ch)
    ct = np.asarray(ct)

    lo = approx_center - approx_width
    hi = approx_center + approx_width
    mask = (ch >= lo) & (ch <= hi)

    x = ch[mask]
    y = ct[mask]

    if len(x) < 10:
        raise ValueError(f"Too few points in window for peak near {approx_center}.")

    # Initial guesses
    C0 = np.median(y)
    A0 = float(np.max(y) - C0)
    mu0 = float(x[np.argmax(y)])
    sigma0 = max(1.0, approx_width / 3)

    p0 = [A0, mu0, sigma0, C0]

    # Weights
    if use_poisson_weights:
        # sigma_y ~ sqrt(N), avoid zero
        sigma_y = np.sqrt(np.maximum(y, 1.0))
    else:
        sigma_y = None

    popt, pcov = curve_fit(
        gaussian, x, y, p0=p0,
        sigma=sigma_y, absolute_sigma=(sigma_y is not None),
        maxfev=20000
    )

    A_hat, mu_hat, sigma_hat, C_hat = popt
    mu_err = float(np.sqrt(np.abs(pcov[1, 1]))) if pcov.shape == (4, 4) else np.nan

    # Plot fit (window + full)
    xs = np.linspace(x.min(), x.max(), 600)
    plt.figure()
    plt.plot(ch, ct, linewidth=1, label="spectrum")
    plt.axvspan(lo, hi, alpha=0.2, label="fit window")
    plt.plot(xs, gaussian(xs, *popt), linewidth=2, label="gaussian fit")
    plt.title(f"Peak fit near ~{approx_center:.1f} ch: mu={mu_hat:.2f}±{mu_err:.2f}")
    plt.xlabel("ADC channel")
    plt.ylabel("Counts")
    plt.legend()
    if save:
        plt.savefig(PLOTS_PEAKS / filename, dpi=200, bbox_inches="tight")
    plt.show()

    return float(mu_hat), float(mu_err), float(sigma_hat), float(A_hat)


def estimate_peak_windows_from_spectrum(ch, ct, n_peaks, ch_min=0, ch_max=None, guard=50):
    """
    Super-simple heuristic: pick top local maxima after smoothing, spaced apart.
    Use only if you *don't* know approximate channels.
    Returns list of approx_centers.
    """
    ch = np.asarray(ch)
    ct = np.asarray(ct)

    if ch_max is None:
        ch_max = ch.max()

    mask = (ch >= ch_min) & (ch <= ch_max)
    x = ch[mask]
    y = ct[mask]

    # Smooth (moving average)
    k = 7
    y_s = np.convolve(y, np.ones(k)/k, mode="same")

    # crude maxima
    candidates = np.argsort(y_s)[::-1]  # descending
    centers = []
    for idx in candidates:
        c = x[idx]
        if all(abs(c - cc) > guard for cc in centers):
            centers.append(float(c))
        if len(centers) >= n_peaks:
            break

    return sorted(centers)
```

```python
# -----------------------------
# HELPER FUNCTIONS: calibration curve fit (energy vs channel)
# -----------------------------

def linear_cal(x, m, b):
    return m * x + b

def fit_calibration(ch_centers, ch_errs, energies_keV, energy_errs_keV,
                    force_through_origin=False):
    """
    Fit E = m*ch + b (or b=0 if force_through_origin).
    Uses uncertainties from channel fits + energy uncertainties.

    Returns dict with m, b, cov, chi2, dof
    """
    x = np.asarray(ch_centers, dtype=float)
    y = np.asarray(energies_keV, dtype=float)

    sx = np.asarray(ch_errs, dtype=float)
    sy = np.asarray(energy_errs_keV, dtype=float)

    # If you have sx, propagate x-uncertainty approximately into y-uncertainty via initial slope.
    # We'll iterate once: start with unweighted fit, then update.
    if force_through_origin:
        def model(x, m): return m * x
        p0 = [y[-1] / x[-1]]
    else:
        model = linear_cal
        p0 = [ (y[-1]-y[0])/(x[-1]-x[0]), y[0] ]

    # first pass weights: sy only
    sigma_y = np.sqrt(np.maximum(sy, 0.0)**2 + 1e-12)
    popt, pcov = curve_fit(model, x, y, p0=p0, sigma=sigma_y, absolute_sigma=True, maxfev=20000)

    # refine with x-uncertainty folded in (approx): σ_eff^2 = sy^2 + (m*sx)^2
    if force_through_origin:
        m = popt[0]
        sigma_eff = np.sqrt(sy**2 + (m*sx)**2 + 1e-12)
        popt, pcov = curve_fit(model, x, y, p0=[m], sigma=sigma_eff, absolute_sigma=True, maxfev=20000)
        m = popt[0]
        b = 0.0
    else:
        m, b = popt
        sigma_eff = np.sqrt(sy**2 + (m*sx)**2 + 1e-12)
        popt, pcov = curve_fit(model, x, y, p0=[m, b], sigma=sigma_eff, absolute_sigma=True, maxfev=20000)
        m, b = popt

    # goodness of fit
    yhat = model(x, *popt)
    resid = y - yhat
    chi2 = np.sum((resid / sigma_eff)**2)
    dof = max(1, len(x) - len(popt))

    return {
        "m": float(m),
        "b": float(b),
        "cov": pcov,
        "chi2": float(chi2),
        "dof": int(dof),
        "sigma_eff": sigma_eff
    }


def plot_calibration_fit(ch_centers, energies_keV, calfit, title, savepath=None):
    x = np.asarray(ch_centers, dtype=float)
    y = np.asarray(energies_keV, dtype=float)

    m, b = calfit["m"], calfit["b"]

    xs = np.linspace(max(0, x.min()*0.9), x.max()*1.05, 400)
    ys = m*xs + b

    plt.figure()
    plt.scatter(x, y)
    plt.plot(xs, ys, linewidth=2)
    plt.xlabel("ADC channel (peak center)")
    plt.ylabel("Energy (keV)")
    plt.title(f"{title}\nE = {m:.6g} ch + {b:.6g} | chi2/dof = {calfit['chi2']:.2f}/{calfit['dof']}")
    if savepath is not None:
        plt.savefig(savepath, dpi=200, bbox_inches="tight")
    plt.show()
```

```python
# -----------------------------
# HELPER FUNCTIONS: peak extraction driver for a given run
# -----------------------------

def extract_peaks_for_known_energies(run_path: Path,
                                    energies_keV, energies_err_keV,
                                    approx_centers_ch, approx_width_ch=80,
                                    bg_path: Path=None,
                                    t_sample_s=15*60, t_bg_s=15*60,
                                    save_peak_plots=True,
                                    tag="run"):
    """
    Loads run spectrum; optionally background subtracts; fits Gaussian peaks at provided approx centers.

    Returns dataframe with columns:
      energy_keV, energy_err_keV, ch_center, ch_err
    """
    ch, ct = load_spectrum(run_path)

    if bg_path is not None:
        bg_ch, bg_ct = load_spectrum(bg_path)
        ct_corr, ct_corr_err = subtract_background(ch, ct, bg_ch, bg_ct,
                                                   t_sample_s=t_sample_s, t_bg_s=t_bg_s)
        ct_use = ct_corr
    else:
        ct_use = ct

    rows = []
    for E, dE, c0 in zip(energies_keV, energies_err_keV, approx_centers_ch):
        mu, mu_err, sig, A = find_peak_gaussian(
            ch, ct_use,
            approx_center=float(c0),
            approx_width=float(approx_width_ch),
            save=save_peak_plots,
            filename=f"{tag}_E{E:.1f}keV.png"
        )
        rows.append({
            "energy_keV": float(E),
            "energy_err_keV": float(dE),
            "ch_center": float(mu),
            "ch_err": float(mu_err),
            "sigma_ch": float(sig),
            "amp": float(A),
        })

    return pd.DataFrame(rows)
```

```python
# -----------------------------
# HELPER FUNCTIONS: daily adjustment model
# -----------------------------

def calibrate_energy(channel, cal):
    """Given channel(s), return energy using cal dict with m,b."""
    return cal["m"] * np.asarray(channel) + cal["b"]

def derive_daily_adjustment_shift_only(base_cal, daily_peak_ch, daily_peak_E):
    """
    Assumes the *slope is stable* and all drift is a channel offset Δch.
    If base: E = m ch + b,
    and daily peak at channel ch_d corresponds to energy E,
    then Δch = (E - b)/m - ch_d   (the amount you add to daily channels)
    """
    m, b = base_cal["m"], base_cal["b"]
    ch_base_expected = (daily_peak_E - b) / m
    delta_ch = ch_base_expected - daily_peak_ch
    return float(delta_ch)

def apply_channel_shift(ch, delta_ch):
    return np.asarray(ch) + delta_ch

def derive_daily_adjustment_full_linear(daily_peak_chs, daily_peak_Es, daily_peak_ch_errs=None, daily_peak_E_errs=None):
    """
    Fit a daily calibration E = m_d ch + b_d from daily reference peaks.
    (If you only have 1 peak, this is ill-posed; use shift-only.)
    """
    if daily_peak_ch_errs is None:
        daily_peak_ch_errs = np.zeros_like(daily_peak_chs, dtype=float)
    if daily_peak_E_errs is None:
        daily_peak_E_errs = np.zeros_like(daily_peak_Es, dtype=float)

    df = fit_calibration(daily_peak_chs, daily_peak_ch_errs, daily_peak_Es, daily_peak_E_errs, force_through_origin=False)
    return df

def blend_two_cals(calA, calB, w):
    """
    Linear blend of two calibrations (for overnight interpolation).
    w=0 -> A, w=1 -> B
    """
    out = dict(calA)
    out["m"] = (1-w)*calA["m"] + w*calB["m"]
    out["b"] = (1-w)*calA["b"] + w*calB["b"]
    return out
```

```python
# -----------------------------
# GENERATE CALIBRATION DATA
# -----------------------------
```
