# It also has functions like mean, std for data analysis:

```python
test_mean = 10
test_std = 2
n_test_points = 100

test_data = np.random.normal(test_mean, test_std, n_test_points) # use numpy's random number sampling from a normal distribution to generate some fake data

mean_value = np.sum(test_data)/n_test_points   # calculating the mean without np.mean

stdev = np.sqrt( 1./(n_test_points-1) * np.sum( (np.mean(test_data)-test_data)**2) )   # calculating the std error without np.std

error_of_mean = stdev/(np.sqrt(n_test_points))


print("The mean: "+ str(np.mean(test_data))+", or "+str(mean_value)) # our mean and numpy's mean should be the same
print("The standard deviation: " +str(np.std(test_data,ddof=1))+", or "+str(stdev)) # note the ddof=1
print("The standard error of the mean: " + str(error_of_mean))
```

# Making a Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

mean = 50
sigma = 15
scores = np.random.normal(mean,sigma,1000)

plt.figure(figsize=(12,7))

plt.hist(scores,bins=30, color='g')

plt.tight_layout()

plt.title("Midterm Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.rcdefaults()
plt.show()
```

# Ploting histogram with errorbars

```python
plt.figure(figsize=(12,7))
counts, edges = np.histogram(scores, bins=20, range=(0, 100))
bin_err = counts**0.5
cbins = (edges[:-1] + edges[1:])/2
plt.errorbar(cbins, counts, yerr=bin_err, fmt="go", label="Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.legend()
```

```python
x5 = np.linspace(0,100,5) # making linearly spaced x values between min and max with n points
x500 = np.linspace(0,100,500) # more smooth

# for plotting purposeses, make a more "continuous" x and y variables, with much finer spacing

y5 = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x5 - mean)**2 / (2 * sigma**2) )
y500 = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x500 - mean)**2 / (2 * sigma**2) )

plt.figure(figsize=(12,7))

plt.plot(x5, y5, linewidth=4, color="r", label="True Values (coarse)")
plt.plot(x500, y500, linewidth=4, color="k", label="True Values (fine)")

plt.hist(scores, bins=30, density=True, color='g', label="Random Sample")
# plt.tight_layout()
plt.title("Midterm Score Distribution")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

# Basic examples of functions:

```python
# start function definitions with the "def " and the name of the function
# inside the parentheses you place the variables which squared takes in
def squared(thingtosquare):
    """This is a documentation string for the function squared, which takes
    in a number (thingtosquare) and outputs its square (thingsquared).
    NB multi-line comments are in triple quotes, and we have to tab carefully in function blocks!"""
    thingsquared = thingtosquare * thingtosquare     # simply squares the input variable, and returning it
    # x*x is generally faster than x**2, not so for higher powers
    return thingsquared
```

```python
# define a function to make a gaussian with input values, used later
def mygauss(x, amp, center, sigma):
    """This is an example gaussian function, which takes in x values, the amplitude (amp),
    the mean of the distributation (center) and the sigma of the Gaussian, and returns the respective y values."""
    y = amp * np.exp(-.5*((x-center)/sigma)**2)
    return y
```

```python
npts = 40  
x = np.linspace(0,10,npts) 
amp = 3.1
center = 5.1
sigma = 1.11
y = mygauss(x, amp, center, sigma)
```

```python
plt.figure(figsize=(12,7))
plt.plot(x, y, 'bo', label='data points')     
plt.text(center, amp, "<-- peak is here",fontsize=16)     # places text at any x/y location on the graph
plt.xlabel('X axis',fontsize=20)
plt.ylabel('Y axis', fontsize=20)
plt.title('A gaussian plot \n with some extras!',fontsize=20)
plt.legend(loc='best')
```

# Now we'll pretend to have some data, with some errors, and do a fitting exercise.

```python
def mygauss_errors(x, amp, center, sigma, noise_amp):
    """Takes in x values, amplitude, center, sigma for a gaussian, as well as a noise amplitude
    which multiplies a random normally distributed (sigma=1,center=0) noise that is added to the gaussian"""
    y = amp * np.exp(-.5*((x-center)/sigma)**2) #+ np.random.randn(len(x)) 
    y += np.random.randn(len(x)) * noise_amp
    return y

# make up some hidden parameters for the noisy Gaussian, we will try to discover them with optimization later
npts, minx, maxx = 15, 1, 10
x = np.linspace(minx, maxx, npts)    
amp, center, sigma = 200, 5.3, 1.11   # made up hidden parameters
noise_amp = .05*amp   # this multiplies a unit normal distribution to simulate noise, larger means increased noise

# get the noisy y values using the above input parameters
y = mygauss_errors(x, amp, center, sigma, noise_amp)

xcont = np.linspace(1,10,1000)
ycont = mygauss(xcont, amp, center, sigma) # also get the true continuous y values

plt.figure(figsize=(12,7))
plt.plot(xcont, ycont, 'r', label='True values (nature)')

plt.errorbar(x, y, yerr=noise_amp, marker='o', linestyle='None', label='Noisy data') #plots points with error 
plt.title('Fake data example with a noisy Gaussian')
plt.legend()
```

# Now that we have some fake data, let's pretend we don't know the parameters and try to find them via Chi squared minimization:

```python
# to fit a function to some data, you will need to minimize (optimize) some statistic
# scipy has a function in it which will take in a model and data and find the best parameters
# for you using a chi-squared minimization
from scipy.optimize import curve_fit

# you can always google for more info about common functions, but python has documentation strings
# that you can access in the notebook by typing

help(curve_fit)

# or also executing "curve_fit?" would pop up a window with the same information
```

```python
# use the curve_fit function to find optimal parameters for the mygauss function, given the data
# (it does better with a good guess for starting parameters, though none or even bad guesses work)
amp_guess, center_guess, sigma_guess = 42, 4.2, .42
params, cov = curve_fit(mygauss, x, y, p0=[amp_guess,center_guess,sigma_guess])

print("These are the best fit amplitude, center, and sigma")
print(params)
print("These are the covariances between those parameters (i.e. diagonal elements are sigma**2 for each param.)")
print(cov)

# these will print out the values and their errors, according to the covariance matrix
print("True amplitude, center, and sigma: " +str(amp)+", "+str(center)+', '+str(sigma))
print("Fitted amplitude, center, and sigma: " +str(params))
print("With estimated errors +/- " +str(np.sqrt(np.diag(cov))))
chi2 = 1./(len(x)-3)*np.sum((y - mygauss(x, *params))**2/(noise_amp)**2)
print('Reduced chi-squared of fit: '+str(chi2))
```

```python
str(list(zip(params,np.diag(cov))))
```

# Now plot all the curves together: the noisy data, the true nature, and the fit

```python
plt.figure(figsize=(10,6))

plt.errorbar(x, y, yerr=noise_amp, marker='o', linestyle='None', label='Noisy data') 

plt.plot(xcont, mygauss(xcont, amp, center, sigma), label='True data (nature)')
fiterrs = np.sqrt(np.diag(cov))

#plt.fill_between(xcont,mygauss(xcont,*params-fiterrs),mygauss(xcont,*params+fiterrs),color='orange',alpha=.5)


plt.plot(xcont, mygauss(xcont, *params), label='Fitted curve', color='r')
plt.axhline(0, color='k', linestyle='--')

plt.xlabel('X axis', fontsize=20)
plt.ylabel('Y axis', fontsize=20)
plt.title('Our result: A noisy Gaussian with fit', fontsize=20)
plt.legend(loc='upper left')
plt.show()
```

# Just for fun... we can repeat all the above, faking the data, and fitting it as many times as we want. Does it return reasonable model fits? This is a way of checking the measurement/modeling process for bias. Hint: lower/raise the number of points (npts) and noise amplitude (noise_amp) variables below to see how e.g. the reduced chi-squared distribution changes

```python
npts, minx, maxx = 100, 0, 9
x = np.linspace(minx,maxx,npts)    # making linearly spaced x values between min and max with n points
amp, center, sigma = 50.1, 5.3, 1.11   # made up hidden parameters
noise_amp = 2   # this multiplies a unit normal distribution to simulate noise, larger means increased noise


# Now just repeat the experiment with all the above true parameters
# and store the resulting model parameters
nExps = 10000    # the number of times to run the expt, 10k takes a few seconds
all_chi2 = np.zeros(nExps)
all_amp, all_cen, all_sig = np.zeros(nExps), np.zeros(nExps), np.zeros(nExps)
all_amp_std, all_cen_std, all_sig_std = np.zeros(nExps), np.zeros(nExps), np.zeros(nExps)

t_start = time.time()   #mark the start time
for i in range(nExps):    # simply loop over repeated "experiments", saving the model parameters each time
    y = mygauss_errors(x, amp, center, sigma, noise_amp)    # fake the data
    params, cov = curve_fit(mygauss, x, y, p0=[amp_guess,center_guess,sigma_guess])    # fit the model
    all_chi2[i] = 1./(len(x)-3)*np.sum((y-mygauss(x,*params))**2/(noise_amp)**2)    # record chi2
    all_amp[i], all_cen[i], all_sig[i] = params    # save parameters
    all_amp_std[i], all_cen_std[i], all_sig_std[i] = np.array([np.sqrt(cov[diag][diag]) for diag in range(3)])

print("took "+str(time.time()-t_start)+" seconds")   #current time minus start time is total time to run
```

# There are lots of things we could now plot... for instance, is the reduced chi2 distribution reasonable?

```python
plt.figure(figsize=(12,7))
plt.hist(all_chi2, bins=100, histtype='step')
plt.xlabel('$\\chi^2_r$', fontsize=20)
plt.title('The $\\chi^2$ of '+str(nExps)+' repeated "measurement simulations"')
plt.axvline(1, color='k')
```

# Looking at the distribution of model fits and residuals:

```python
nPlots = 50
true_y = mygauss(xcont, amp, center, sigma)
plt.figure(figsize=(10,4))
for i in range(nPlots):
    plt.subplot(1,2,1)
    fit_y = mygauss(xcont, all_amp[i], all_cen[i], all_sig[i])
    plt.plot(xcont, fit_y, 'b.', alpha=.01)    #plot the best fit model for this expt
    plt.subplot(122)
    plt.plot(xcont, (fit_y-true_y)/amp*100., 'k.', alpha=.01)    #plot the residuals


plt.ylabel('% difference')
plt.title('Residuals of best fit model minus truth')
plt.axhline(0, color='k', linestyle='-', label='residuals')
plt.xlabel('x')
plt.legend()
plt.subplot(121)
plt.plot(xcont, ycont, 'k', linewidth=2, label='true nature') #plot the true nature being measured
plt.title('Spread of best fit model \n in repeated experiments')
plt.legend()
```
